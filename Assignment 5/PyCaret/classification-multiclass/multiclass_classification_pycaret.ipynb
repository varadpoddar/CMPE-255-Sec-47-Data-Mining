# PyCaret Multiclass Classification — Iris

Dataset: `iris.csv`. Target: `species`.

## Environment
Using Python 3.11 with `pycaret==3.3.0` inside the shared `.venv`. Notebook assumes data CSVs are present in `../data/` and GPU is available; if not, PyCaret will fall back to CPU.

```python
import pandas as pd
from pycaret.classification import *

csv_path = "../data/iris.csv"
df = pd.read_csv(csv_path)
df.rename(columns={'Species':'species'}, inplace=True, errors='ignore')
print(df.head())
print(df['species'].value_counts())
```

```python
s = setup(
    data=df,
    target='species',
    session_id=42,
    train_size=0.8,
    fold=5,
    normalize=True,
    use_gpu=False,
    log_experiment=True,
    experiment_name='pycaret_multiclass_iris',
)
compare_models();
```

```python
best = compare_models()
tuned = tune_model(best)
plot_model(tuned, plot='confusion_matrix', save=True)
plot_model(tuned, plot='class_report', save=True)
```

```python
from pycaret.classification import *
best = create_model("lightgbm")  # or "catboost"/"rf"/"xgboost"
tuned = tune_model(best)
plot_model(tuned, plot="feature", save=True)
```

```python
final_best = finalize_model(tuned)
preds = predict_model(final_best)
print(preds.head())
save_model(final_best, 'multiclass_iris_model')
```
