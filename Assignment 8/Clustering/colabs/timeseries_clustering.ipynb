# Time-Series Clustering
Using tslearn (KShape/TimeSeriesKMeans) on synthetic series. Run in Colab.

```
# Install
!pip install -q tslearn matplotlib
```

```
import numpy as np, matplotlib.pyplot as plt
from tslearn.clustering import TimeSeriesKMeans
from tslearn.datasets import CachedDatasets
# synthetic: extract from Trace dataset sample
X_train, y_train, _, _ = CachedDatasets().load_dataset('Trace')
X = X_train[:100]
model = TimeSeriesKMeans(n_clusters=3, metric='dtw', max_iter=10, random_state=42)
labels = model.fit_predict(X)
for c in range(3):
    for ts in X[labels==c][:5]:
        plt.plot(ts.ravel(), alpha=0.5)
    plt.title(f'Cluster {c}')
    plt.show()
```
