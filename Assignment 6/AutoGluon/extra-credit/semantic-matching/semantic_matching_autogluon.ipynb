# AutoGluon Semantic Matching (tiny synthetic)

```python
from autogluon.multimodal import MultiModalPredictor
import pandas as pd
from pathlib import Path

time_limit = 30
try:
    HERE = Path(__file__).resolve().parent
except NameError:
    HERE = Path('.')
```

## Text-to-Text Matching

```python
text_pairs = [
    ('A cat on a mat', 'A cat is sitting on a mat', 1),
    ('The sky is blue', 'The ocean is deep', 0),
    ('He reads a book', 'A man is reading a novel', 1),
    ('She likes apples', 'He hates oranges', 0),
]
df_text = pd.DataFrame(text_pairs, columns=['query','response','label'])
predictor_text = MultiModalPredictor(label='label', query='query', response='response', match_label='label', eval_metric='accuracy', path=str(HERE / 'ag_semantic_models' / 'text2text'))
predictor_text.fit(df_text, time_limit=time_limit)
text_metrics = predictor_text.evaluate(df_text)
text_metrics
```
