```python
from autogluon.multimodal import MultiModalPredictor
import pandas as pd
from pathlib import Path

time_limit = 30
# synthetic reviews with numeric features
rows = [
    {'review':'Great battery life and camera', 'price':799, 'rating':5, 'label':1},
    {'review':'Terrible build quality', 'price':299, 'rating':1, 'label':0},
    {'review':'Average phone with decent screen', 'price':499, 'rating':3, 'label':0},
    {'review':'Excellent performance and display', 'price':899, 'rating':5, 'label':1},
    {'review':'Poor battery and slow', 'price':399, 'rating':2, 'label':0},
    {'review':'Fantastic value and features', 'price':599, 'rating':4, 'label':1},
    {'review':'Mediocre camera but cheap', 'price':199, 'rating':3, 'label':0},
    {'review':'Loving the speed and battery', 'price':699, 'rating':5, 'label':1},
]
df = pd.DataFrame(rows)
train_df = df.sample(frac=0.75, random_state=42)
val_df = df.drop(train_df.index)
train_df.head()
```

```python
predictor = MultiModalPredictor(label='label', path='Assignment 6/AutoGluon/extra-credit/multimodal-tabular/ag_multimodal_models', problem_type='binary', eval_metric='accuracy')
predictor.fit(train_df, tuning_data=val_df, time_limit=time_limit)
metrics = predictor.evaluate(val_df)
metrics
```
