# AutoGluon Time Series (indepth/chronos quick check)

```
from autogluon.timeseries import TimeSeriesPredictor
import pandas as pd
from pathlib import Path

try:
    HERE = Path(__file__).resolve().parent
except NameError:
    HERE = Path('.').resolve()
data_path = HERE / 'data' / 'series.csv'
df = pd.read_csv(data_path, parse_dates=['timestamp'])
train_data = df.iloc[:-12]
test_data = df.iloc[-12:]
train_data.head()
```

```
predictor = TimeSeriesPredictor(
    prediction_length=12,
    path='Assignment 6/AutoGluon/extra-credit/timeseries/ag_timeseries_models',
)
predictor.fit(train_data, time_limit=60)
leaderboard = predictor.leaderboard()
leaderboard.head()
```

```
preds = predictor.predict(train_data, known_covariates=None)
print(preds.head())
metrics = predictor.evaluate(test_data, predictions=preds)
metrics
```
