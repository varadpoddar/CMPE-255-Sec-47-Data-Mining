# AutoGluon Image Classification (synthetic)

Small run on 8 random images across two classes.

```python
from autogluon.multimodal import MultiModalPredictor
import pandas as pd
from pathlib import Path

time_limit = 30
try:
    HERE = Path(__file__).resolve().parent
except NameError:
    HERE = Path('.')
img_dir = HERE / 'data' / 'train'
rows = []
for cls_dir in img_dir.iterdir():
    if cls_dir.is_dir():
        for img in cls_dir.glob('*.png'):
            rows.append({'image': str(img), 'label': cls_dir.name})
df = pd.DataFrame(rows)
train_df = df.sample(frac=0.75, random_state=42)
val_df = df.drop(train_df.index)
train_df.head()
```

```python
predictor = MultiModalPredictor(label='label', path=str(HERE / 'ag_image_models'))
predictor.fit(train_df, tuning_data=val_df, time_limit=time_limit)
metrics = predictor.evaluate(val_df)
metrics
```
